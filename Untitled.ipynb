```python
import pandas as pd
from nltk.stem.porter import PorterStemmer

import pickle
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on = 'title')
```

```python
# genres , id , keywords , title , overview , cast , crew
movies = movies[['movie_id' , 'title' , 'overview', 'genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.dropna(inplace = True)     # drop nan values rows because we have huge data and nan have few
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy" , }, {"id": 878, "name": "Science Fiction"}]
# ["Action" , "Adventure" , "Fantasy" , "Science Fiction"]
# to convert '[list]' into [list] use ast.literal_eval(obj1)
def convert_genres(obj1):
    li = []
    for i in ast.literal_eval(obj1):
        li.append(i['name'])
    return li

def convert_cast(obj1):
    li = []
    counter = 0
    for i in ast.literal_eval(obj1):
        if counter !=3:
            li.append(i['name'])
            counter+=1
        else:
            break
    return li

def Convert_crew(obj):
    li = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            li.append(i['name'])
            break
    return li
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
movies['keywords'] = movies['keywords'].apply(convert_genres)
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(Convert_crew)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk   # uses to convert samilear words into same words like [love,lower,lovely] -> [love,love,love]
PS = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(PS.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer      # use to convert words into form of vectors
cv = CountVectorizer(max_features = 5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity         # use to map similarity words with other movies like it creates (4806,4806) dimention matrics
```

```python
similarity = cosine_similarity(vectors)      # hear it map vectors and generate a similarity score
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]     # hear we get top 5 similarity score of a first matrics 
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
